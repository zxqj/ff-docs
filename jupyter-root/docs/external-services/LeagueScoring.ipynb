# Scoring Settings
## Available Settings
### Sleeper
There are 13, 8, 16, 20, 40, and 6 configurable settings in the Sleeper UI for Passing, Rushing, Receiving, Kicking, Defense, and Special Teams (Team), respectively.  There are 3 "miscellaneous" and 14 "bonus" settings.  Finally, there are 6 Special Teams (Player) settings, and 21 IDP settings.  All of the bonus and miscellaneous settings apply to teams with only offensive player roles.  I believe that, of the 6 Special Teams (Player) settings, 4 of them are probably regular point contributors to players occupying offensive fantasy roster slots (`player td`, `fumble recovery`, `punt return yards`, `kick return yards` - though the first two are somewhat rare).  `forced fumble` and `solo tackle` would be exceedingly rare and would mostly only happen if the returner fumbled the ball and did something about it himself.

This gives 13+8+16+20+40+6+3+14+4 + (very rarely)2 = 118-120 settings for rosters with no defensive players, and 141 settings total.

Points listed immediately after the setting, not in parentheses, are the defaults for PPR.  Otherwise, the setting contributes 0 points by default.

Note that the position the player has does not matter except if specified in the setting.

Broken down another way, the settings counts are
* OFF  - 37
* K/ST - 32
* DEF  - 40
* IDP  - 21
* Other- 17

#### OFF (37)
##### Passing (13)
* Common (4)
    * points per yard .04
    * td 4
    * 1st down
    * 2pt conv 2
* Volume (2)
    * completed
    * pass attempts
* Demerit (4)
    * intercepted -1
    * pick6 thrown
    * incompleted pass
    * qb sacked
* Yardage/TD Bonus (3)
    * 40+ bonus
    * 40+ td bonus
    * 50+ td bonus

##### Rushing (8)
* Common (4)
    * points per yard .1
    * td 4
    * 1st down
    * 2pt conv 2
* Volume (1)
    * att
* Yardage/TD Bonus (3)
    * 40+ bonus
    * 40+ td bonus
    * 50+ td bonus

##### Receiving (16)
* Common (4)
    * points per yard .1
    * td 6
    * 1st down
    * 2pt conv 2
* Volume (1)
    * reception 1
* Yardage/TD Bonus (8)
    * yard bonus settings (0-4, 5-9, 10-19, 20-29, 30-39, 40+)
    * yard + td bonus settings (40+, 50+)
* Position Bonus (3)
    * (RB, WR, TE)

#### K/ST (32)

##### Kicking (20)
* (4) FG made/missed, PAT made/missed: 0/-1, 1/-1
* (2) Points per yd, points per yd missed: 0, 0
* (14) yardage settings: 7 for made fgs, and 7 for missed.  0-19, 20-29, 30-39, 40-49, 50-59, 60+, and then a 50+ bonus.  For made field goals, the defaults are 3, 3, 3, 4, 5, 6, and 0, respectively.  The default values for the missed settings are all zero.  Effectively, $$\text{points}(\text{yards}) = max(3,\lfloor\frac{\text{yards}}{10}\rfloor)$$

##### Special Teams (Team) (6)
* td 6
* forced fumble 1
* fumble recovery 1
* solo tackle
* points per punt return yard
* points per kick return yard

##### Special Teams (Player) (6)
* The same as for Special Teams (Team) (but for individual players)

#### Defense (40)
* Will document these comprehensively later if necessary
* All values set to zero by default, except for 13 settings
    * (1) td 6
    * (6) points allowed
        * points allowed (<20) - defaults given by $$\text{fantasypoints}(\text{pointsallowed}) = 10 - 3 \times \lfloor\frac{\text{pointsallowed}}{7}\rfloor $$
        * points allowed (>20)
            * 28-34  -1
            * 35+    -4
    * (6) Other
        * Forced fumble 1
        * Sack 1
        * Fumble recovery 2
        * Interception 2
        * Safety 2
        * Blocked Kick 2

#### IDP (21)
* Points per(13)
    * Tackle
    * Assisted Tackle
    * QB Hit
    * Solo Tackle
    * TFL
    * Safety
    * Blocked Kick
    * Forced Fumble
    * Fumble Recovery
    * Pass Defended
    * Sack
    * Interception
    * Touchdown
* Points per yard
    * Sack
    * Interception Return
    * Fumble Return
* Per game bonus
    * 10+ Tackles
    * 2+ Sacks
    * 3+ Passes Defended
* Per play bonus
    * 50+yd fumble return && 1+ TD
    * 50+yd interception return && 1+ TD
      

#### Other (17)

##### Miscellaneous (3)
* Fumble
* Fumble Lost -2
* Fumble Recovery TD 6

##### Bonus (14)
* Game yardage bonuses (8)
    * 100-199 and 200+: for rushing, receiving, or combined rushing/receiving
    * 300-399 and 400+ for passing
* Game volume bonuses (2):
    * 20+ carries
    * 25+ pass completions
* Position Bonuses (4): QB, TE, WR, RB
