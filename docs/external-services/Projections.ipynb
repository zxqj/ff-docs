# Projections

These are the specifics of APIs that produce player stats projections for future NFL games

## Sleeper
This one is not part of their published API, but is requested by their web application.

A token is sent in the authorization header; however, just hitting the URL with `curl` and no header returned a result

`https://api.sleeper.com/projections/nfl/2024/13?season_type=regular&position[]=DEF&position[]=K&position[]=QB&position[]=RB&position[]=TE&position[]=WR&order_by=pts_ppr`

To query season, just chop off last component of path
